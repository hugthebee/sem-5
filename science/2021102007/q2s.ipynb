```python
# -- PART A -- #

import matplotlib.pyplot as plt
import numpy as np

# we will multiply and divide by c to obtain the term v/c
m = 10
c = 3 * (10 ** 8)

v = np.linspace(0,0.99,500) # generates 500 numbers between 0 and 1
non_relative = m*c*v # velocity is written in terms of c

# for relative mometum, we get p = mcx/(1-x^2)^(1/2)
v2 = np.square(v)
relative = m*c*v/np.sqrt(1-v2)

plt.figure()
plt.plot(v,relative,label = "Relativistic momentum")
plt.grid("on")
plt.xlabel("v/c")
plt.ylabel("Momentum (kgm/s)")
plt.plot(v,non_relative,label = "Non - relativistic momentum")

plt.legend(["Relativistic momentum", "Non-relativistic momentum"])
# print(result)
```

```python
 # -- PART B -- #

non_relative = 0.5*m*v2*c*c # velocity is written in terms of c

# for relative ke, we get 
relative = (m*c*c/np.sqrt(1-v2)) - m*c*c

plt.figure()
plt.plot(v,relative,label = "Relativistic Kinetic Energy")
plt.grid("on")
plt.xlabel("v/c")
plt.ylabel("Kinetic Energy (J)")
plt.plot(v,non_relative,label = "Linear kinetic Energy")

plt.legend(["Relativistic Kinetic Energy", "Linear Kinetic Energy"])
# print(result)
```
